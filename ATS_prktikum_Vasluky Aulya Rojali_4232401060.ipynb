```python
1. # Meminta input suhu dari pengguna
suhu = float(input("Masukkan suhu air pendingin (°C): "))

# Percabangan kondisi suhu
if suhu <= 90:
    print("STATUS: Normal")
elif suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")
```

Penjelasan:

float(input(...)) digunakan agar input bisa berupa bilangan desimal.

Struktur if-elif-else memeriksa kondisi secara berurutan:

- Jika suhu <= 90, maka status Normal.

- Jika tidak memenuhi syarat pertama tapi suhu < 100, maka status Warning.

- Jika tidak memenuhi keduanya (artinya suhu >= 100), maka ALERT: SHUTDOWN.

```python
3. # Data vibrasi dari 5 unit mesin
vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

# Loop untuk memeriksa setiap nilai vibrasi
for i in range(len(vibrasi_mesin)):
    if vibrasi_mesin[i] > 2.0:
        print(f"Warning: Unit {i+1} vibrasi tinggi!")
```

```python
5. # Inisialisasi variabel
suhu_target = 85
suhu_saat_ini = 60
kenaikan_suhu = 5

print(f" Sistem Pemanas Dimulai")
print(f"Suhu Awal: {suhu_saat_ini}°C")
print(f"Suhu Target: {suhu_target}°C")
print("-" * 30)

# Looping while akan berjalan selama suhu_saat_ini kurang dari suhu_target (85°C)
while suhu_saat_ini < suhu_target:
    # Naikkan suhu saat ini sebesar 5 derajat
    suhu_saat_ini += kenaikan_suhu

    # Cetak suhu saat ini setelah kenaikan
    # Gunakan min() untuk memastikan suhu tidak melebihi target jika kenaikan 5 derajat
    # membuatnya melompat melebihinya, meskipun untuk kasus ini tidak terjadi.
    suhu_cetak = min(suhu_saat_ini, suhu_target)

    # Untuk mencetak suhu yang naik per iterasi
    print(f"Iterasi: Suhu saat ini adalah {suhu_cetak}°C")

# Setelah loop selesai
print("-" * 30)
print(f" Pemanasan Selesai. Suhu telah mencapai {suhu_saat_ini}°C (Target {suhu_target}°C).")
```

```python
 7. #Percabangan / Branching: Klasifikasi Beban Listrik

# Menerima input dari pengguna
# Gunakan float() untuk memungkinkan nilai desimal
kapasitas_maksimum = float(input("Masukkan Kapasitas Maksimum (kW): "))
beban_terpakai = float(input("Masukkan Beban Terpakai Saat Ini (kW): "))

print("-" * 35)

# Hitung persentase pemakaian
persentase = (beban_terpakai / kapasitas_maksimum) * 100

print(f"Persentase Pemakaian: {persentase:.2f}%")
print("-" * 35)

# Percabangan untuk menentukan status beban
if persentase <= 60:
    status = "Beban Ringan"
elif persentase > 60 and persentase <= 85:
    status = "Beban Sedang"
else:  # Ini mencakup persentase > 85
    status = "Beban Berat"

# Tampilkan hasil
print(f"Status Beban: **{status}**")
```

Penjelasan:Input Pengguna: Program menerima dua nilai numerik dari pengguna: kapasitas_maksimum dan beban_terpakai.Perhitungan Persentase: Persentase pemakaian dihitung menggunakan rumus:$$\text{Persentase} = \left(\frac{\text{beban\_terpakai}}{\text{kapasitas\_maksimum}}\right) \times 100$$Percabangan if: Kondisi pertama diperiksa. Jika persentase <= 60, maka statusnya adalah "Beban Ringan". Jika kondisi ini benar, sisa percabangan diabaikan.Percabangan elif: Jika kondisi pertama salah, program melanjutkan ke elif. Kondisi ini menggunakan operator and untuk memeriksa apakah persentase > 60 DAN persentase <= 85. Jika benar, statusnya adalah "Beban Sedang".Percabangan else: Jika semua kondisi di atas (if dan elif) salah, ini berarti persentase pasti di atas $85$. Oleh karena itu, statusnya ditetapkan sebagai "Beban Berat" tanpa perlu pengecekan kondisi lebih lanjut.

```python
#SOAL NO 9
suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]

print("=== Pembacaan Suhu Boiler ===")
for i, suhu in enumerate(suhu_boiler, start=1):
    print(f"Titik {i}: {suhu} °C", end=" ")
    if suhu > 100:
        print(" PERINGATAN: Suhu Melebihi Batas Aman!")
    else:
        print("- Aman")

rata = sum(suhu_boiler) / len(suhu_boiler)
print(f"\nRata-rata suhu boiler: {rata:.2f} °C")

if rata <= 90:
    status = "Dingin"
elif rata <= 100:
    status = "Normal"
else:
    status = "Panas"

print(f"Kondisi boiler saat ini: {status}")
```

PENJELASAN PROGRAM
Program diawali dengan sebuah list bernama suhu_boiler, yang berisi data hasil pengukuran suhu pada lima titik berbeda di dalam boiler. Tipe data list dipilih karena mampu menyimpan beberapa nilai suhu sekaligus dalam satu variabel, dan mudah diolah menggunakan perulangan.

Perulangan for dan Enumerasi
Program menggunakan perulangan for untuk menampilkan setiap suhu beserta urutan titiknya: Fungsi enumerate() digunakan agar setiap nilai suhu memiliki nomor urut otomatis (mulai dari 1), sehingga lebih mudah dibaca di hasil keluaran.

Percabangan if di dalam loop
Di dalam loop, digunakan percabangan if untuk memeriksa apakah suhu melebihi 100°C: Jika suhu di atas 100°C, program menampilkan peringatan untuk operator agar segera melakukan pemeriksaan atau pendinginan.

Menghitung Rata-rata Suhu
Setelah semua data ditampilkan, program menghitung rata-rata suhu keseluruhan menggunakan fungsi sum() dan len(): Nilai rata-rata ini memberikan gambaran umum tentang kondisi termal boiler secara keseluruhan.

Klasifikasi Kondisi Boiler Berdasarkan nilai rata-rata suhu, program menggunakan struktur if–elif–else untuk menentukan kondisi boiler: ≤ 90°C → Boiler dalam kondisi dingin, mungkin efisiensi rendah.

90–100°C → Normal, aman dan stabil.

100°C → Panas, perlu perhatian lebih karena risiko overheat.



```python
# Challange
# Program: Simulasi Listrik Rumah Tangga (Sederhana)

def energi_kwh(daya_watt, jam_per_hari, hari=30):
    """Hitung energi dalam kWh untuk periode hari (default 30 hari)."""
    return (daya_watt * jam_per_hari * hari) / 1000.0

def tampilkan_format(nama, daya, jam, kwh):
    """Format tampilan tabel hasil."""
    print(f"{nama:15} | {daya:4} W | {jam:5.1f} jam/hari | {kwh:7.2f} kWh/bulan")

def to_float_or_default(s, default):
    """Jika input kosong kembalikan default, jika tidak konversi ke float."""
    if s.strip() == "":
        return default
    try:
        return float(s)
    except:
        return default

perangkat = [
    ("Kulkas", 150),          # kulkas 150 W
    ("Kipas", 50),            # kipas 50 W
    ("Lampu x4 (15W)", 15*4), # 4 lampu 15W = total 60W
    ("TV", 100),              # TV 100 W
    ("Casan HP x4", 10*4)     # 4 charger HP, masing-masing 10 W
]

default_jam = {
    "Kulkas": 16.0,           # kulkas nyala rata-rata 16 jam/hari
    "Kipas": 6.0,
    "Lampu x4 (15W)": 5.0,
    "TV": 4.0,
    "Casan HP x4": 3.0        # cas HP rata-rata 3 jam/hari
}

print("=== Perhitungan Konsumsi Listrik Rumah Tangga (Sederhana) ===")
print("Masukkan jam pemakaian per-hari untuk tiap alat.")
print("Tekan ENTER untuk gunakan nilai default.\n")

jam_per_hari = {}
for nama, daya in perangkat:
    inp = input(f"Jam/hari untuk {nama} (default {default_jam[nama]}): ")
    jam = to_float_or_default(inp, default_jam[nama])
    jam_per_hari[nama] = jam


tarif = input("\nMasukkan tarif listrik Rp/kWh (tekan ENTER pakai default Rp1120 - tarif Batam): ")
tarif = to_float_or_default(tarif, 1120.0)

print("\n=== Rincian Konsumsi Bulanan (30 hari) ===")
print("Nama              | Daya  | Jam/hari | kWh/bulan")
print("------------------------------------------------")
total_kwh = 0.0
for nama, daya in perangkat:
    jam = jam_per_hari[nama]
    kwh = energi_kwh(daya, jam, hari=30)
    tampilkan_format(nama, daya, jam, kwh)
    total_kwh += kwh


biaya = total_kwh * tarif
print("------------------------------------------------")
print(f"Total konsumsi bulanan: {total_kwh:.2f} kWh")
print(f"Estimasi biaya bulan  : Rp {biaya:,.0f}")

if total_kwh < 100:
    kat = "Rendah"
elif total_kwh <= 300:
    kat = "Sedang"
else:
    kat = "Tinggi"

print(f"Kategori konsumsi     : {kat}")

if kat == "Rendah":
    print("Saran: Pemakaian listrik Anda efisien ")
elif kat == "Sedang":
    print("Saran: Gunakan alat elektronik seperlunya agar lebih hemat ")
else:
    print("Sara8n: Konsumsi tinggi! Pertimbangkan mengganti peralatan hemat energi ")

print("\nSelesai. Terima kasih telah menggunakan program simulasi listrik rumah tangga.")
```
